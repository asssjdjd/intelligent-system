```python
# ============================================================
# Visualization of Student Marks across Subjects
# Author: Phm Thanh Long - B22DCCN505
# Description:
#     This program visualizes student performance across 
#     multiple subjects using two chart types:
#         (1) Multi-line plot for trend comparison
#         (2) Grouped bar chart for absolute comparison
#     Additional features:
#         - Annotations on bar chart
#         - Computation of subject averages
#         - Export of figures
# ============================================================

import matplotlib.pyplot as plt
import numpy as np

# ------------------------------------------------------------
# Step 1: Data preparation
# ------------------------------------------------------------
students = ['Ann', 'Joe', 'Mark']
subjects = ['Math', 'Physics', 'Chemistry', 'English', 'CS']

marks = [
    [80, 75, 90, 85, 70],   # Ann
    [60, 65, 70, 72, 68],   # Joe
    [90, 88, 85, 92, 95]    # Mark
]

marks_array = np.array(marks)

# Compute average marks per subject
subject_means = marks_array.mean(axis=0)

print("=== Average Marks by Subject ===")
for subj, mean in zip(subjects, subject_means):
    print(f"{subj:10s}: {mean:.2f}")

# ------------------------------------------------------------
# Step 2: Multi-line plot
# ------------------------------------------------------------
plt.figure(figsize=(8, 5))
markers = ['o', 's', '^']  # custom markers per student

for i, student in enumerate(students):
    plt.plot(
        subjects,
        marks[i],
        marker=markers[i],
        label=student,
        linewidth=2
    )

plt.title("Student Marks - Multiple Line Plot", fontsize=14, weight="bold")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()
plt.savefig("student_marks_lineplot.png", dpi=300)
plt.show()

# ------------------------------------------------------------
# Step 3: Grouped bar chart
# ------------------------------------------------------------
x = np.arange(len(subjects))
width = 0.25

plt.figure(figsize=(8, 5))
for i, student in enumerate(students):
    bars = plt.bar(
        x + width * i,
        marks[i],
        width=width,
        label=student
    )
    # Add annotation on top of each bar
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            height + 1,
            f"{height}",
            ha="center", va="bottom", fontsize=9
        )

plt.xticks(x + width, subjects)
plt.title("Student Marks - Grouped Bar Chart", fontsize=14, weight="bold")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.ylim(0, 110)  # make space for labels above bars
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("student_marks_barchart.png", dpi=300)
plt.show()

```
