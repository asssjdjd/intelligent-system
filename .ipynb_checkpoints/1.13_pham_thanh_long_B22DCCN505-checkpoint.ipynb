```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
# Pham Thanh Long - B22DCCN505
# Dữ liệu
heights = np.array([1.6, 1.65, 1.7, 1.73, 1.8]).reshape(-1,1)
weights = np.array([60, 65, 72.3, 75, 80]).reshape(-1,1)

# Huấn luyện mô hình Linear Regression
model = LinearRegression()
model.fit(heights, weights)

# Vẽ dữ liệu gốc
plt.title('weights plotted against heights')
plt.xlabel('heights in meters')
plt.ylabel('weights in kilograms')
plt.plot(heights, weights, 'k.', label='Data points')
plt.axis([1.5, 1.85, 50, 90])
plt.grid(True)

# Vẽ đường hồi quy
plt.plot(heights, model.predict(heights), color='r', label='Regression line')
plt.legend()
plt.show()
```
