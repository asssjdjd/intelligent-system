```python
# ============================================================
# Linear Regression Example: Predicting Weight from Height
# Author: Pháº¡m Thanh Long - B22DCCN505
# Description:
#     This script demonstrates a simple linear regression model
#     to predict body weight (kg) from height (m).
#     It visualizes the data points and regression line,
#     and prints the regression equation + RÂ² score.
# ============================================================

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# Step 1: Data preparation
# ------------------------------------------------------------
heights = np.array([1.6, 1.65, 1.7, 1.73, 1.8]).reshape(-1, 1)
weights = np.array([60, 65, 72.3, 75, 80]).reshape(-1, 1)

# ------------------------------------------------------------
# Step 2: Train Linear Regression model
# ------------------------------------------------------------
model = LinearRegression()
model.fit(heights, weights)

slope = model.coef_[0][0]
intercept = model.intercept_[0]
r2 = r2_score(weights, model.predict(heights))

print("=== Regression Model Parameters ===")
print(f"Slope (coefficient): {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"Equation: weight = {slope:.4f} * height + {intercept:.4f}")
print(f"RÂ² score: {r2:.4f}")

# ------------------------------------------------------------
# Step 3: Visualization
# ------------------------------------------------------------
plt.figure(figsize=(7, 5))

# Original data points
plt.scatter(heights, weights, color='blue', s=60, label='Data points')

# Regression line
height_range = np.linspace(1.55, 1.85, 100).reshape(-1, 1)
plt.plot(height_range, model.predict(height_range), 
         color='red', linewidth=2, label='Regression line')

# Formatting
plt.title("Weight vs Height - Linear Regression", fontsize=14, weight="bold")
plt.xlabel("Height (m)", fontsize=12)
plt.ylabel("Weight (kg)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# Annotation with equation
eq_text = f"y = {slope:.2f}x + {intercept:.2f}\nRÂ² = {r2:.2f}"
plt.text(1.55, 85, eq_text, fontsize=10, color="red", 
         bbox=dict(facecolor='white', alpha=0.6, edgecolor='red'))

plt.tight_layout()
plt.savefig("linear_regression_height_weight.png", dpi=300)
plt.show()
```
