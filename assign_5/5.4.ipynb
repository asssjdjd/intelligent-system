```python
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import tensorflow as tf
import matplotlib.pyplot as plt

# ===== 1. Load dataset =====
df = pd.read_csv("filmReview_1.csv")
texts = df["review_text"].astype(str).tolist()
labels = df["rating"].values

# multiclass one-hot
lb = LabelBinarizer()
y = lb.fit_transform(labels)

# ===== 2. Tokenize & sequence =====
max_words = 30000
maxlen = 256
tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
X = pad_sequences(sequences, maxlen=maxlen, padding="post")

# ===== 3. Train Word2Vec =====
tokenized_sentences = [t.split() for t in texts]
w2v = Word2Vec(sentences=tokenized_sentences, vector_size=200, window=5, min_count=2, workers=4, epochs=5)

# Build embedding matrix
word_index = tokenizer.word_index
num_words = min(max_words, len(word_index) + 1)
embedding_matrix = np.zeros((num_words, 200))
for word, i in word_index.items():
    if i >= num_words: 
        continue
    if word in w2v.wv:
        embedding_matrix[i] = w2v.wv[word]

# ===== 4. Train/val split =====
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state=42, stratify=labels)

# ===== 5. Build CNN model =====
def build_cnn_model():
    inp = tf.keras.layers.Input(shape=(maxlen,))
    emb = tf.keras.layers.Embedding(input_dim=num_words,
                                    output_dim=200,
                                    weights=[embedding_matrix],
                                    input_length=maxlen,
                                    trainable=False)(inp)
    x = tf.keras.layers.Conv1D(128, 5, activation="relu")(emb)
    x = tf.keras.layers.GlobalMaxPooling1D()(x)
    x = tf.keras.layers.Dense(128, activation="relu")(x)
    x = tf.keras.layers.Dropout(0.5)(x)
    out = tf.keras.layers.Dense(y.shape[1], activation="softmax")(x)
    model = tf.keras.Model(inp, out)
    return model

cnn_model = build_cnn_model()
cnn_model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
cnn_model.summary()

# ===== 6. Train CNN =====
history_cnn = cnn_model.fit(X_train, y_train,
                            validation_data=(X_val, y_val),
                            epochs=5,
                            batch_size=64)

# ===== 7. Build LSTM model =====
def build_lstm_model():
    inp = tf.keras.layers.Input(shape=(maxlen,))
    emb = tf.keras.layers.Embedding(input_dim=num_words,
                                    output_dim=200,
                                    weights=[embedding_matrix],
                                    input_length=maxlen,
                                    trainable=False)(inp)
    x = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128))(emb)
    x = tf.keras.layers.Dense(128, activation="relu")(x)
    x = tf.keras.layers.Dropout(0.5)(x)
    out = tf.keras.layers.Dense(y.shape[1], activation="softmax")(x)
    model = tf.keras.Model(inp, out)
    return model

lstm_model = build_lstm_model()
lstm_model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
lstm_model.summary()

# ===== 8. Train LSTM =====
history_lstm = lstm_model.fit(X_train, y_train,
                              validation_data=(X_val, y_val),
                              epochs=5,
                              batch_size=64)

# ===== 9. Build SimpleRNN model =====
def build_rnn_model():
    inp = tf.keras.layers.Input(shape=(maxlen,))
    emb = tf.keras.layers.Embedding(input_dim=num_words,
                                    output_dim=200,
                                    weights=[embedding_matrix],
                                    input_length=maxlen,
                                    trainable=False)(inp)
    x = tf.keras.layers.SimpleRNN(128)(emb)
    x = tf.keras.layers.Dense(128, activation="relu")(x)
    x = tf.keras.layers.Dropout(0.5)(x)
    out = tf.keras.layers.Dense(y.shape[1], activation="softmax")(x)
    model = tf.keras.Model(inp, out)
    return model

rnn_model = build_rnn_model()
rnn_model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
rnn_model.summary()

# ===== 10. Train RNN =====
history_rnn = rnn_model.fit(X_train, y_train,
                            validation_data=(X_val, y_val),
                            epochs=5,
                            batch_size=64)


                          

def plot_history(histories, titles):
    plt.figure(figsize=(14,6))

    # Plot accuracy
    plt.subplot(1,2,1)
    for h, label in zip(histories, titles):
        plt.plot(h.history["accuracy"], label=f"{label} Train")
        plt.plot(h.history["val_accuracy"], label=f"{label} Val")
    plt.title("Model Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()

    # Plot loss
    plt.subplot(1,2,2)
    for h, label in zip(histories, titles):
        plt.plot(h.history["loss"], label=f"{label} Train")
        plt.plot(h.history["val_loss"], label=f"{label} Val")
    plt.title("Model Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()

    plt.tight_layout()
    plt.show()

# Gọi visualize cho cả 3 mô hình
plot_history(
    [history_cnn, history_lstm, history_rnn],
    ["CNN", "BiLSTM", "SimpleRNN"]
)


from transformers import AutoTokenizer, TFAutoModel

# ===== 11. Load pretrained BERT from HuggingFace =====
tokenizer_bert = AutoTokenizer.from_pretrained("bert-base-uncased")
bert_model = TFAutoModel.from_pretrained("bert-base-uncased")

# Hàm lấy embedding BERT [CLS]
def get_bert_embeddings(texts, batch_size=32):
    all_embeddings = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i+batch_size]
        inputs = tokenizer_bert(batch_texts, padding=True, truncation=True, return_tensors="tf", max_length=256)
        outputs = bert_model(inputs)
        cls_emb = outputs.last_hidden_state[:,0,:]  # lấy vector [CLS], shape (batch, 768)
        all_embeddings.append(cls_emb.numpy())
    return np.vstack(all_embeddings)

print("Extracting BERT embeddings...")
X_bert = get_bert_embeddings(texts)

# ===== 12. Train/val split cho BERT =====
X_train_bert, X_val_bert, y_train_bert, y_val_bert = train_test_split(X_bert, y, test_size=0.15, random_state=42, stratify=labels)

# ===== 13. Build CNN on BERT =====
def build_cnn_bert():
    inp = tf.keras.layers.Input(shape=(768,))
    x = tf.keras.layers.Reshape((768,1))(inp)
    x = tf.keras.layers.Conv1D(128, 5, activation="relu")(x)
    x = tf.keras.layers.GlobalMaxPooling1D()(x)
    x = tf.keras.layers.Dense(128, activation="relu")(x)
    x = tf.keras.layers.Dropout(0.5)(x)
    out = tf.keras.layers.Dense(y.shape[1], activation="softmax")(x)
    model = tf.keras.Model(inp, out)
    return model

cnn_bert = build_cnn_bert()
cnn_bert.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
cnn_bert.summary()

history_cnn_bert = cnn_bert.fit(X_train_bert, y_train_bert,
                                validation_data=(X_val_bert, y_val_bert),
                                epochs=5,
                                batch_size=32)

# ===== 14. Build LSTM on BERT =====
def build_lstm_bert():
    inp = tf.keras.layers.Input(shape=(768,))
    x = tf.keras.layers.RepeatVector(1)(inp)        # biến vector 768 thành chuỗi length=1
    x = tf.keras.layers.LSTM(128)(x)
    x = tf.keras.layers.Dense(128, activation="relu")(x)
    x = tf.keras.layers.Dropout(0.5)(x)
    out = tf.keras.layers.Dense(y.shape[1], activation="softmax")(x)
    model = tf.keras.Model(inp, out)
    return model

lstm_bert = build_lstm_bert()
lstm_bert.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
lstm_bert.summary()

history_lstm_bert = lstm_bert.fit(X_train_bert, y_train_bert,
                                  validation_data=(X_val_bert, y_val_bert),
                                  epochs=5,
                                  batch_size=32)

# ===== 15. Build RNN on BERT =====
def build_rnn_bert():
    inp = tf.keras.layers.Input(shape=(768,))
    x = tf.keras.layers.RepeatVector(1)(inp)
    x = tf.keras.layers.SimpleRNN(128)(x)
    x = tf.keras.layers.Dense(128, activation="relu")(x)
    x = tf.keras.layers.Dropout(0.5)(x)
    out = tf.keras.layers.Dense(y.shape[1], activation="softmax")(x)
    model = tf.keras.Model(inp, out)
    return model

rnn_bert = build_rnn_bert()
rnn_bert.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
rnn_bert.summary()

history_rnn_bert = rnn_bert.fit(X_train_bert, y_train_bert,
                                validation_data=(X_val_bert, y_val_bert),
                                epochs=5,
                                batch_size=32)

# ===== 16. Visualize BERT-based models =====
plot_history(
    [history_cnn_bert, history_lstm_bert, history_rnn_bert],
    ["CNN-BERT", "LSTM-BERT", "RNN-BERT"]
)
```
