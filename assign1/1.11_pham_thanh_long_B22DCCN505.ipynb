```python
# ============================================================
# Salary Distribution Analysis by Gender
# Author: Phạm Thanh Long - B22DCCN505
# Description:
#     This script visualizes salary distribution across genders 
#     using swarm plot, complemented with boxplot overlay for 
#     better understanding of central tendency and spread.
# ============================================================

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ------------------------------------------------------------
# Step 1: Load dataset
# ------------------------------------------------------------
data = pd.read_csv('salary.csv')

# Kiểm tra nhanh thống kê mô tả
print("=== Descriptive Statistics by Gender ===")
print(data.groupby("gender")["salary"].describe())

# ------------------------------------------------------------
# Step 2: Create swarm plot + boxplot
# ------------------------------------------------------------
plt.figure(figsize=(7, 5))
sns.set_style("whitegrid")

# Boxplot (nền)
sns.boxplot(
    x="gender", y="salary", data=data,
    showcaps=True, boxprops={'facecolor':'None'},
    showfliers=False, whiskerprops={'linewidth':1.5}
)

# Swarmplot (trên boxplot)
sns.swarmplot(
    x="gender",
    y="salary",
    hue="gender",       # chỉ định hue = gender
    data=data,
    size=5,
    alpha=0.8,
    palette="Set2",
    legend=False        # không cần legend vì trục X đã hiển thị giới tính
)
# ------------------------------------------------------------
# Step 3: Customize chart
# ------------------------------------------------------------
plt.title("Salary Distribution by Gender", fontsize=14, weight="bold")
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Salary", fontsize=12)

# Thêm giá trị trung bình mỗi nhóm
group_means = data.groupby("gender")["salary"].mean()
for i, (gender, mean_val) in enumerate(group_means.items()):
    plt.text(i, mean_val, f"Mean: {mean_val:.0f}", 
             ha="center", va="bottom", fontsize=9, weight="bold", color="red")

plt.tight_layout()
plt.savefig("salary_distribution.png", dpi=300)
plt.show()
```
