```python
# ============================================================
# Linear Regression Example
# Author: Phạm Thanh Long - B22DCCN505
# Description:
#     This script demonstrates simple linear regression 
#     using scikit-learn. It fits a model, makes predictions, 
#     and visualizes both original data and regression line.
# ============================================================

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# Step 1: Data preparation
# ------------------------------------------------------------
X = np.array([50, 60, 65, 70, 75, 80, 85]).reshape(-1, 1)
y = np.array([2.5, 3, 3.5, 3.8, 4, 4.5, 5])

# ------------------------------------------------------------
# Step 2: Train the model
# ------------------------------------------------------------
model = LinearRegression()
model.fit(X, y)

print("=== Regression Model Parameters ===")
print(f"Coefficient (slope): {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
print(f"Equation: y = {model.coef_[0]:.4f} * X + {model.intercept_:.4f}")

# ------------------------------------------------------------
# Step 3: Predictions
# ------------------------------------------------------------
X_new = np.array([55, 68, 76, 90]).reshape(-1, 1)
predictions = model.predict(X_new)

print("\n=== Predictions ===")
for xi, yi in zip(X_new.ravel(), predictions):
    print(f"X = {xi:>3} → Predicted y = {yi:.2f}")

# ------------------------------------------------------------
# Step 4: Visualization
# ------------------------------------------------------------
plt.figure(figsize=(8, 5))

# Original data
plt.scatter(X, y, color='blue', label='Original data', s=70)

# Regression line (drawn across the full range of X)
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_range_pred = model.predict(X_range)
plt.plot(X_range, y_range_pred, color='green', linewidth=2, label='Regression line')

# Predicted points
plt.scatter(X_new, predictions, color='red', marker='x', s=100, label='Predicted points')

# Add annotation for each prediction
for xi, yi in zip(X_new.ravel(), predictions):
    plt.text(xi, yi + 0.05, f"{yi:.2f}", ha='center', fontsize=9, color='red')

# Formatting
plt.xlabel("X", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.title("Linear Regression Model and Predictions", fontsize=14, weight="bold")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("linear_regression_predictions.png", dpi=300)
plt.show()

```
